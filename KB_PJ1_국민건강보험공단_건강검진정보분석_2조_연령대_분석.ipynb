#### 비만 / 복부 비만

```python
# 복부비만

def belly(x):
    # 남성
    if x['성별코드'] == 1:
        if x['허리둘레'] >= 90:
            return '해당'
        else:
            return '비해당'
            
    # 여성
    else:
        if x['허리둘레'] >= 85:
            return '해당'
        else:
            return '비해당'
        

df['복부비만'] = df.apply(belly, axis = 1)
df.head()
```

```python
# 비만

def overweight(x):
    h = x['신장(5Cm단위)']
    h *= 0.01
    h = h ** 2
    w = x['체중(5Kg 단위)']
    bmi = w / h
    
    if bmi <= 18.5:
        return '저체중'
    elif bmi <= 25:
        return '정상'
    elif bmi <= 30:
        return '과체중'
    else:
        return '비만'
    
df['비만'] = df.apply(overweight, axis = 1)
df.head()
```

```python
# 빈혈 함수
# 성별코드 1:남자, 2:여자
def 빈혈(x):
    if x['성별코드'] == 1:
        if x['혈색소'] <= 13:
            return '해당'
        else:
            return '비해당'
    
    else:
        if x['혈색소'] <= 12:
            return '해당'
        else:
            return '비해당'

df['빈혈'] = df.apply(빈혈,axis =1)
df.head(1)
```

```python
con1 = (df['식전혈당(공복혈당)']<= 120)
con2 = (df['식전혈당(공복혈당)']>120)&(df['식전혈당(공복혈당)']<126)
con3 = (df['식전혈당(공복혈당)']>=126)

df.loc[con1, '당뇨'] = "정상"
df.loc[con2, '당뇨'] = "질환 위험"
df.loc[con3, '당뇨'] = "질환 의심"
df
```

```python
def blood_pressure(x):
    sbp = x['수축기 혈압']
    dpb = x['이완기 혈압']
    
    if dpb <= 60:
        return '저혈압'
        
    elif dpb >= 95:
        return '고혈압'
    
    else:
        if sbp <= 100:
            return '저혈압'
        
        elif sbp >= 145:
            return '고혈압'
        
        else:
            return '정상'

df['혈압'] = df.apply(blood_pressure, axis=1)
```

```python
def liver(x):
    ast = x['(혈청지오티)AST']
    alt = x['(혈청지오티)ALT']
    
    if ast <= 32:
        if alt <= 32:
            return '정상'
        elif alt >= 46:
            return '질환의심'
        else:
            return '질환위험'
        
    elif ast >= 51:
        return '질환의심'
    
    else:
        return '질환위험'
        
df['간 기능 이상'] = df.apply(liver, axis = 1)
```
